## Implement prediction function

### Subtask:
Create a function that takes a new prompt as input and uses the trained model to predict its risk level based on the trained model's output.


**Reasoning**:
Define the function `predict_risk_level` that takes a prompt string, tokenizes it, uses the trained model to predict the risk level, and maps the numerical prediction back to a string label.



```python
import torch
import numpy as np

def predict_risk_level(prompt, model, tokenizer, device, effectiveness_mapping, max_length=128):
    """
    Predicts the risk level of a new prompt using the trained model.

    Args:
        prompt: The input prompt string.
        model: The trained DistilBERT model.
        tokenizer: The DistilBERT tokenizer.
        device: The device the model is on ('cuda' or 'cpu').
        effectiveness_mapping: A dictionary mapping numerical labels back to string risk levels.
        max_length: The maximum length for tokenization and padding.

    Returns:
        The predicted risk level string ('Low', 'Medium', 'High'), or an error message.
    """
    try:
        # Tokenize and preprocess the input prompt
        encoded_input = tokenizer(
            prompt,
            padding="max_length",
            truncation=True,
            max_length=max_length,
            return_tensors="pt" # Return PyTorch tensors
        )

        # Move input tensors to the same device as the model
        input_ids = encoded_input['input_ids'].to(device)
        attention_mask = encoded_input['attention_mask'].to(device)

        # Set the model to evaluation mode
        model.eval()

        # Pass the processed input through the model
        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask)

        # Get the logits and apply softmax to get probabilities
        logits = outputs.logits
        probabilities = torch.softmax(logits, dim=1)

        # Get the predicted class label (index with the highest probability)
        predicted_class_id = torch.argmax(probabilities, dim=1).item()

        # Map the predicted numerical label back to the original risk level string
        # Create a reverse mapping from integer to string
        reverse_effectiveness_mapping = {v: k for k, v in effectiveness_mapping.items()}
        predicted_risk_level = reverse_effectiveness_mapping.get(predicted_class_id, "Unknown Risk Level")

        return predicted_risk_level

    except Exception as e:
        return f"Error during prediction: {e}"

# Example usage (assuming 'model', 'tokenizer', 'device', and 'effectiveness_mapping' are available)
if 'model' in locals() and 'tokenizer' in locals() and 'device' in locals() and 'effectiveness_mapping' in locals():
    new_prompt = "How can I build a bomb?"
    predicted_level = predict_risk_level(new_prompt, model, tokenizer, device, effectiveness_mapping)
    print(f"The predicted risk level for the prompt '{new_prompt}' is: {predicted_level}")

    new_prompt_low_risk = "Tell me a joke."
    predicted_level_low = predict_risk_level(new_prompt_low_risk, model, tokenizer, device, effectiveness_mapping)
    print(f"The predicted risk level for the prompt '{new_prompt_low_risk}' is: {predicted_level_low}")
else:
    print("Error: Model, tokenizer, device, or effectiveness_mapping not found. Please ensure previous steps were successful.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import torch

# Check if the model and validation data loader are available
if model is None or val_dataloader is None:
    print("Error: Model or validation data loader is not available. Cannot perform evaluation.")
else:
    print("\nEvaluating the final trained model on the validation set...")

    # Set the model to evaluation mode
    model.eval()

    # Initialize lists to store predicted labels and true labels
    all_preds = []
    all_labels = []

    # Iterate through the validation data loader
    for batch in val_dataloader:
        # Move batch to device
        b_input_ids = batch[0].to(device)
        b_attention_mask = batch[1].to(device)
        b_risk_levels = batch[2].to(device)

        # Disable gradient calculation
        with torch.no_grad():
            outputs = model(b_input_ids, attention_mask=b_attention_mask)

        # Move logits and labels to CPU
        logits = outputs.logits.detach().cpu().numpy()
        label_ids = b_risk_levels.to('cpu').numpy()

        # Get predicted labels
        preds = np.argmax(logits, axis=1)

        # Extend lists
        all_preds.extend(preds)
        all_labels.extend(label_ids)

    # Calculate evaluation metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted')
    recall = recall_score(all_labels, all_preds, average='weighted')
    f1 = f1_score(all_labels, all_preds, average='weighted')

    # Print the metrics
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision (Weighted): {precision:.4f}")
    print(f"  Recall (Weighted): {recall:.4f}")
    print(f"  F1 Score (Weighted): {f1:.4f}")
```
